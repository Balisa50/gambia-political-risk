# 03, Exploratory data analysis

Visualise the corpus before any modelling. All outputs save to `outputs/eda/`.

```python
import pandas as pd
import matplotlib.pyplot as plt
from src.config import CLEAN_CSV, EDA_DIR

df = pd.read_csv(CLEAN_CSV, parse_dates=['date'])
print(len(df), 'articles')
```

## Article volume over time

```python
monthly = df.set_index('date').resample('MS').size()
fig, ax = plt.subplots(figsize=(11, 4))
monthly.plot(ax=ax, color='#00f0ff')
ax.set_title('Articles per month'); ax.set_ylabel('count')
plt.tight_layout(); plt.savefig(EDA_DIR / 'monthly_volume.png'); plt.show()
```

## Top 50 most frequent words

```python
from collections import Counter
words = Counter()
for toks in df['tokens_str'].fillna('').str.split():
    words.update(toks)
top = pd.Series(dict(words.most_common(50))).sort_values()
top.tail(50).plot.barh(figsize=(8, 12)); plt.title('Top 50 words')
plt.tight_layout(); plt.savefig(EDA_DIR / 'top_words.png'); plt.show()
```
