## Plot with event annotations

```python
import matplotlib.pyplot as plt
import pandas as pd
from src.config import CHARTS_DIR, ELECTION_DATES, PRI_CSV

pri = pd.read_csv(PRI_CSV, parse_dates=['week'])
fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(pri['week'], pri['pri'], color='#00f0ff', lw=1.5)
for d, label in ELECTION_DATES:
    ax.axvline(pd.to_datetime(d), color='#ff5577', linestyle='--', alpha=0.6)
    ax.text(pd.to_datetime(d), 95, label, rotation=90, fontsize=8, color='#ff5577')
ax.set_title('Gambia Political Risk Index, weekly')
ax.set_ylabel('PRI (0-100)'); ax.set_ylim(0, 100)
plt.tight_layout(); plt.savefig(CHARTS_DIR / 'political_risk_index.png'); plt.show()
```
